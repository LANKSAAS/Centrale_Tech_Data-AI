## **3) Créer la Sauvegarde (git commit -m)**

Maintenant que les  fichiers sont dans la **Staging Area**, on va l'enregistrer **définitivement dans le repertoire local** (le Repository (créer grâce à  init)).
  
      Chaque commit doit avoir un message expliquant ce qui a été fait.

      -m (pour message)


```python
# git commit -m "message de commit ici"

# exemple : ->(sur cmd) git commit -m "mon premier commit"
```

## **Résumer (Partie 1)**

On a ce cycle de travail : 
   - Initialisation du répôt local dans lequel va s'enregistrer les **commits**(les sauvegardes) et les **logs** (historique)   : --> **git init**

   - Pour voir où j'en suis (état du répertoire local) : **git status** 

   - Pour ajouter des ajouter fichiers dans le répertoire (**Staying Area**) : --> **git add . ou git add <nom_fichier>**
   
   - Pour sauvegarder dans le répertoire : --> **git commit  -m 'Mon message ici'**

## **4) Création de branches**

* **Voir la liste de branches** 

```python
# git  branch

#output : * master  : le * indique la branche courante/ la branche sur laquelle on se trouve actuellement
```

* **Créer une nouvelle branche** 
       
                git branch <nom_de_branche>

```python
# git branch  <nom_de_branche>

#Disclaimer: cette commande crée la branche, mais ne nous y bascule pas automatiquement, pour cela il faut faire git checkout <nom_de_branche>
#on rester toujours sur la branche master ou main 

# git checkout <nom_de_branche>  ou git switch <nom_de_branche>
# permet de basculer sur la branche que l'on veut celle crée
```

**Astuce**  :  **créer une branche et basculer sur la branche crée**

                git checkout -b <nom_de_la branche>
                ou 
                git switch -c <nom_de_la branche>

                c (pour create)

Une fois sur cette branche, on peut faire les git add et git commit normalement. Ces sauvegardes n'existeront que sur cette branche.

* **Fusionner des branches**

          git merge

**Disclaimer** :
  
         Règle d'or du Merge : Toujours ce placer dans sur la branche qui va recevoir les modifications.

```python
# exemple 

# on creer une branche de travail pour ajouter une nouvelle fonctionnalité, on l'appelle feature1
# git switch -c feature1
# on ajoute une nouvelle fonctionnalité dans notre projet, on commit les changements
# git add .
# git commit -m "ajout de la fonctionnalité 1"
# on bascule sur la branche main pour ajouter une autre fonctionnalité
# git switch main
# On merge la branche feature1 dans main
# git merge feature1
```

* **Supprimer une branche** 

          git branch -d <nom_de_branche>

## **5) Lien avec Github**

Se trouver dans le même dossier que là où a été fait les git commit.

* **connecter son Pc à son dépôt github**


      git remote add origin
      remote add : ajouter un dépôt distant 
      origin = le nom standard qu'on donne au dépôt distant Principal

```python
#exemple

# git remote add origin <url_de_votre_depot_git>

#(sur cmd) git remote add origin https://github.com/votre_utilisateur/votre_depot.git
```

* **Envoyer son code sur Github**


        Pour envoyer (pousser) tes commits locaux vers GitHub, on utilise la commande : push 
        git push ou git push -u  (-u : crée un lien permanent)

```python
# git push -u origin main
```

* **Récupérer le code (Clone et Pull)**


      git clone


```python
# git clone https://github.com/ton-pseudo/mon-premier-projet.git
```

* **Mettre à jour ton code** 


      git pull
